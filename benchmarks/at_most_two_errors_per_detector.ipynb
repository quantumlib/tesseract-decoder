```python
from plot import plot

from utils import create_label

from run_and_collect import run_and_collect
```

```python
# RUN COLOR CODE BENCHMARK

color_file_path = '../testdata/colorcodes/r=11,d=11,p=0.002,noise=uniform,c=superdense_color_code_Z,q=181,gates=all.stim'
color_shots = 200

color_data = run_and_collect(color_file_path, color_shots)
```

```python
color_data_at_most_two_errors = run_and_collect(color_file_path, color_shots, at_most_two_errors_per_detector=True)
```

```python
# RUN SURFACE CODE BENCHMARK
surface_file_path = '../testdata/surfacecodes/r=11,d=11,p=0.002,noise=si1000,c=surface_code_Z,q=241,gates=cz.stim'
surface_shots = 500

surface_data = run_and_collect(surface_file_path, surface_shots)
```

```python
surface_at_most_two_errors = run_and_collect(surface_file_path, surface_shots, at_most_two_errors_per_detector=True)
```

```python
# PLOT COLOR AND SURFACE CODE RESULTS: SPEED

plot_data = [
    {"label": create_label(color_data['code_type'], color_data['r'], color_data['p'], color_data['shots']), "before": color_data["seconds"], "after": color_data_at_most_two_errors["seconds"]},
    {"label": create_label(surface_data['code_type'], surface_data['r'], surface_data['p'], surface_data['shots']), "before": surface_data["seconds"], "after": surface_at_most_two_errors["seconds"]},
]

legend_labels = [
    "Any Errors Per Detector",
    "At Most Two Errors Per Detector", 
]

plot(
    data=plot_data,
    title="Speed in Color and Surface Codes With and Without Using '--at-most-two-errors-per-detector'",
    title2="Decoding Time (Seconds)",
    legend=legend_labels,
    size=(16, 8) 
)
```

```python
# PLOT COLOR AND SURFACE CODE RESULTS: ACCURACY

plot_data = [
    {"label": create_label(color_data['code_type'], color_data['r'], color_data['p'], color_data['shots']), "before": color_data["low_confidence"], "after": color_data_at_most_two_errors["low_confidence"]},
    {"label": create_label(surface_data['code_type'], surface_data['r'], surface_data['p'], surface_data['shots']), "before": surface_data["low_confidence"], "after": surface_at_most_two_errors["low_confidence"]},
]

legend_labels = [
    "Any Errors Per Detector",
    "At Most Two Errors Per Detector", 
]

plot(
    data=plot_data,
    title="Accuracy in Color and Surface Codes With and Without Using '--at-most-two-errors-per-detector'",
    title2="Accuracy (Low Confidence Count)",
    legend=legend_labels,
    size=(16, 8) 
)
```
